```python
import pandas as pd

# Load dataset
df = pd.read_csv('/content/Healthcare Providers.csv')

# Inspect the first few rows to understand the data
df.head()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Drop rows with missing values
df = df.dropna()

# Drop columns with more than 30% missing values
df = df.dropna(thresh=int(0.7*len(df)), axis=1)
print(df)
```

```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns
print("Numerical columns:", numerical_columns)
```

```python
# identify categorical columns
categorical_cols = df.select_dtypes(include=['object','category']).columns
print(categorical_cols)
```

```python
#Preprocessing Based on Column Type
# Nominal Columns (No ordering, like "Gender", "Country")
categorical_cols_nominal = ['Gender of the Provider', 'City of the Provider']  # Define nominal columns
df_nominal = pd.get_dummies(df[categorical_cols_nominal], drop_first=True)
print(df_nominal)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance
le = LabelEncoder()

# List of categorical columns to encode
categorical_cols_to_encode = ['Last Name/Organization Name of the Provider',
       'First Name of the Provider', 'Middle Initial of the Provider',
       'Credentials of the Provider', 'Gender of the Provider',
       'Entity Type of the Provider', 'Street Address 1 of the Provider',
       'Street Address 2 of the Provider', 'City of the Provider',
       'State Code of the Provider', 'Country Code of the Provider',
       'Provider Type', 'Medicare Participation Indicator', 'Place of Service',
       'HCPCS Code', 'HCPCS Description', 'HCPCS Drug Indicator',
       'Number of Services', 'Number of Medicare Beneficiaries',
       'Number of Distinct Medicare Beneficiary/Per Day Services',
       'Average Medicare Allowed Amount', 'Average Submitted Charge Amount',
       'Average Medicare Payment Amount',
       'Average Medicare Standardized Amount']

# Loop through each categorical column and apply Label Encoding
for col in categorical_cols_to_encode:
    df[col] = le.fit_transform(df[col])
```

```python
#Ordinal Columns (Ordered categories, like "Education Level")
from sklearn.preprocessing import OrdinalEncoder

# Define ordinal columns
categorical_cols_ordinal = ['Number of Medicare Beneficiaries', 'Number of Distinct Medicare Beneficiary/Per Day Services']

ordinal_encoder = OrdinalEncoder()
df_ordinal = df[categorical_cols_ordinal].copy()

# Apply ordinal encoding
df_ordinal[categorical_cols_ordinal] = ordinal_encoder.fit_transform(df_ordinal[categorical_cols_ordinal])
print(df_ordinal)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler instance
scaler = MinMaxScaler()

# Fit and transform the selected columns and store the result in separate columns
df[['National Provider Identifier_scaled', 'Zip Code of the Provider_scaled']] = scaler.fit_transform(df[['National Provider Identifier', 'Zip Code of the Provider']])

# If you need a single 'numerical_column', you can create it based on the scaled columns:
# For example, you could average the scaled values:
# df['numerical_column'] = (df['National Provider Identifier_scaled'] + df['Zip Code of the Provider_scaled']) / 2

# Or you could choose one of the scaled columns as the 'numerical_column':
# df['numerical_column'] = df['National Provider Identifier_scaled']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['numerical_column1', 'numerical_column2']] = scaler.fit_transform(df[['National Provider Identifier','Zip Code of the Provider']])
```
