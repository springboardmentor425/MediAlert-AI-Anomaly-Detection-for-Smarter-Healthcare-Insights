```python
# Convert all numerical columns to numeric, setting invalid parsing to NaN
numerical_columns = [
    'Number of Services', 'Number of Medicare Beneficiaries', 
    'Number of Distinct Medicare Beneficiary/Per Day Services', 
    'Average Medicare Allowed Amount', 'Average Submitted Charge Amount', 
    'Average Medicare Payment Amount', 'Average Medicare Standardized Amount'
]

for column in numerical_columns:
    # Convert to numeric and set invalid parsing as NaN
    cleaned_df[column] = pd.to_numeric(cleaned_df[column], errors='coerce')
# Display info to verify the changes
print("\nDataset Info After Converting Columns to Numeric:")
print(cleaned_df.info())
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler,StandardScaler,OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
# List of nominal columns
nominal_columns = ['Credentials of the Provider','Gender of the Provider','Entity Type of the Provider','City of the Provider','State Code of the Provider',
                   'Country Code of the Provider','Provider Type','Medicare Participation Indicator','Place of Service','HCPCS Description','HCPCS Drug Indicator']

# Loop through each nominal column
for col in nominal_columns:
    unique_count = cleaned_df[col].nunique()  # Get the number of unique values in the column
    print(f"Column '{col}' has {unique_count} unique values.")
    
    # If the column has more than 10 unique values, apply frequency encoding
    if unique_count > 10:
        print(f"Applying frequency encoding to column: {col}")
        freq_map = cleaned_df[col].value_counts().to_dict()  # Create a frequency map
        cleaned_df[col] = cleaned_df[col].map(freq_map)  # Map the frequency values to the column
    else:
        # Otherwise, apply one-hot encoding
        print(f"Applying one-hot encoding to column: {col}")
        one_hot = pd.get_dummies(cleaned_df[col], prefix=col)  # Create one-hot encoded columns
        cleaned_df = pd.concat([cleaned_df, one_hot], axis=1)  # Add the one-hot columns to the DataFrame
        cleaned_df.drop(columns=[col], inplace=True)  # Drop the original column
```

```python
# Display the updated DataFrame info
print("\nThe updated DataFrame info")
print(cleaned_df.info())
```

```python
# Fill missing values for numerical columns with their mean
cleaned_df.fillna(cleaned_df.mean(), inplace=True)

# Convert all boolean columns to integers (True = 1, False = 0)
boolean_columns = cleaned_df.select_dtypes(include='bool').columns  # Find all boolean columns

for col in boolean_columns:
    cleaned_df[col] = cleaned_df[col].astype(int)  # Convert boolean to integer (True -> 1, False -> 0)

# Display the updated DataFrame info after filling NaNs and converting booleans
print("\nUpdated DataFrame info after filling missing values and converting boolean columns to integers:")
print(cleaned_df.info())
```
