```python
import pandas as pd

# Load the dataset
file_path = 'Dl_project/Healthcare Providers.csv'
dataset = pd.read_csv(file_path)

# Display 
dataset.head(), dataset.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
import numpy as np

# Sample data for encoding 

ordinal_column = 'Entity Type of the Provider'
nominal_column = 'Gender of the Provider'
frequency_column = 'City of the Provider'
```

```python
# Ordinal Encoding
ordinal_data = dataset[[ordinal_column]].copy()
ordinal_encoder = OrdinalEncoder()
ordinal_data[ordinal_column + '_Encoded'] = ordinal_encoder.fit_transform(ordinal_data)

# Display results
ordinal_data.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd

# Sample data
data = {'Category': ['Low', 'Medium', 'High', 'Medium', 'Low']}
df = pd.DataFrame(data)

# Define the categories
categories_order = [['Low', 'Medium', 'High']]

# Apply Ordinal Encoding
encoder = OrdinalEncoder(categories=categories_order)
df['Category_Encoded'] = encoder.fit_transform(df[['Category']])

print(df)
```

```python
# Frequency Encoding
frequency_data = dataset[[frequency_column]].copy()
frequency_map = frequency_data[frequency_column].value_counts().to_dict()
frequency_data[frequency_column + '_Encoded'] = frequency_data[frequency_column].map(frequency_map)

# Display results
frequency_data.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Load the dataset
data = pd.read_csv('Dl_Project/Healthcare Providers.csv')
```

```python
# Step 1: Inspect the dataset
print("Dataset Shape:", data.shape)
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
# Step 2: Handle missing values
# Fill missing categorical values with the mode
data['Gender of the Provider'].fillna(data['Gender of the Provider'].mode()[0], inplace=True)
```

```python
# Fill missing numerical values with the median
data['Zip Code of the Provider'].fillna(data['Zip Code of the Provider'].median(), inplace=True)
```

```python
# Verify existing columns
print("Columns in dataset:", data.columns)

# Drop columns with excessive missing values (only if they exist)
columns_to_drop = ['Street Address 2 of the Provider', 'Middle Initial of the Provider']
columns_to_drop = [col for col in columns_to_drop if col in data.columns]
if columns_to_drop:
    data.drop(columns=columns_to_drop, inplace=True)
    print(f"Dropped columns: {columns_to_drop}")
else:
    print("No columns to drop from the specified list.")
```

```python
# Step 3: Remove duplicates
data.drop_duplicates(inplace=True)
```
