```python
df = df.drop(columns=['Street Address 2 of the Provider'])
```

```python
#For numerical columns, fill with mean or median:
# df['Column_Name'].fillna(df['Column_Name'].mean(), inplace=True)
```

```python
#For categorical columns, fill with mode or "Unknown"
df['First Name of the Provider']=df['First Name of the Provider'].fillna('Unknown')
df['Middle Initial of the Provider']=df['Middle Initial of the Provider'].fillna('Unknown')
df['Credentials of the Provider']=df['Credentials of the Provider'].fillna('Unknown')
df['Gender of the Provider']=df['Gender of the Provider'].fillna('Unknown')

```

```python
print(df.isnull().sum())
```

```python
df
```

```python
# Remove $ and convert to float
df['Average Medicare Allowed Amount'] = df['Average Medicare Allowed Amount'].str.replace('$', '').str.replace(',', '').astype(float)
df['Average Submitted Charge Amount'] = df['Average Submitted Charge Amount'].str.replace('$', '').str.replace(',', '').astype(float)
df['Average Medicare Payment Amount'] = df['Average Medicare Payment Amount'].str.replace('$', '').str.replace(',', '').astype(float)
df['Average Medicare Standardized Amount'] = df['Average Medicare Standardized Amount'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
for i in df.columns:
    print(i,'----->',df[i].unique())
    print()
```

```python
# ordinal
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()

# Apply encoding to the column
df['Medicare Participation Indicator'] = le.fit_transform(df['Medicare Participation Indicator'])
df['HCPCS Drug Indicator'] = le.fit_transform(df['HCPCS Drug Indicator'])
```

```python
# print(df['Medicare Participation Indicator'])
print(df['HCPCS Drug Indicator'])
```

```python
# Convert categorical columns using LabelEncoder
categorical_columns = ['Gender of the Provider', 'Provider Type', 'Entity Type of the Provider', 'Place of Service']
for col in categorical_columns:
   df[col] = LabelEncoder().fit_transform(df[col])
```

```python
df.dtypes
```

```python
# Convert object columns to numeric
df['Number of Services'] = pd.to_numeric(df['Number of Services'], errors='coerce')
df['Number of Medicare Beneficiaries'] = pd.to_numeric(df['Number of Medicare Beneficiaries'], errors='coerce')
df['Number of Distinct Medicare Beneficiary/Per Day Services'] = pd.to_numeric(df['Number of Distinct Medicare Beneficiary/Per Day Services'], errors='coerce')
```

 if the column 'State Code of the Provider' contains values like ['NY', 'CA', 'TX'], it generates new binary columns:

```python
# One hot Encoding
df = pd.get_dummies(df, columns=['State Code of the Provider', 'Country Code of the Provider'], drop_first=True)
```

```python
import matplotlib.pyplot as plt
print("from the above diagram at x-axies amount zero the frequency is high as nearly 100000")
# Plot histogram for 'Average Medicare Payment Amount'
plt.figure(figsize=(8, 6))
df['Average Medicare Payment Amount'].hist(bins=20, edgecolor='black')
plt.title('Distribution of Average Medicare Payment Amount')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

# Plot boxplot for 'Average Medicare Payment Amount' to detect outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=df['Average Medicare Payment Amount'])
plt.title('Boxplot of Average Medicare Payment Amount')
plt.show()
```

```python
# Plot correlation heatmap for the numerical columns
correlation_matrix = df[['Average Medicare Payment Amount', 'Average Medicare Allowed Amount', 'Zip Code of the Provider', 'Medicare Participation Indicator']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plot scatter plot to explore the relationship between 'Average Medicare Allowed Amount' and 'Average Medicare Payment Amount'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Average Medicare Allowed Amount', y='Average Medicare Payment Amount', data=df)
plt.title('Scatter Plot: Average Medicare Allowed Amount vs. Average Medicare Payment Amount')
plt.xlabel('Average Medicare Allowed Amount')
plt.ylabel('Average Medicare Payment Amount')
plt.show()
```
