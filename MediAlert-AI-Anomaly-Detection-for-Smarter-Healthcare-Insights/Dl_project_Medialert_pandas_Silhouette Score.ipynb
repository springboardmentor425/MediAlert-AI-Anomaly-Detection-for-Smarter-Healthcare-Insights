```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Load the dataset from the uploaded file
file_path = 'Dl_project/Healthcare Providers.csv'

# Read the CSV file into a Pandas DataFrame
dataset = pd.read_csv(file_path)
# Step 1: Select relevant numerical features
columns_to_use = [
    'Number of Services',
    'Number of Medicare Beneficiaries',
    'Average Medicare Payment Amount'
]
data_cleaned = dataset[columns_to_use].copy()

# Step 2: Handle missing values and type conversion
# Convert columns to numeric
for col in columns_to_use:
    data_cleaned[col] = pd.to_numeric(data_cleaned[col], errors='coerce')

# Fill missing values with column mean
data_cleaned = data_cleaned.fillna(data_cleaned.mean())

# Step 3: Normalize the data
scaler = StandardScaler()
processed_data = scaler.fit_transform(data_cleaned)

# Step 4: Apply K-Means clustering
n_clusters = 3  # Example: 3 clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(processed_data)

# Step 5: Calculate Silhouette Score
score = silhouette_score(processed_data, cluster_labels)
print(f"Silhouette Score for {n_clusters} clusters: {score}")
```
