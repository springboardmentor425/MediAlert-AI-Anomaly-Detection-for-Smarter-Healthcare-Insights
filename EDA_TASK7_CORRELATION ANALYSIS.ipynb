```python
# Assuming your dataset is loaded
data = pd.read_csv('Healthcare Providers.csv')
```

```python
# Display available column names
print(data.columns)
```

```python
# Select numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['float64', 'int64'])
```

```python
# If no numeric columns, notify the user
if numeric_data.empty:
    print("No numeric columns found for correlation analysis.")
else:
    # Compute the correlation matrix
    correlation_matrix = numeric_data.corr()

    # Display the correlation matrix
    print("Correlation Matrix:")
    print(correlation_matrix)
```

```python
# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Pair plot (if at least two numeric columns are present)
if len(numeric_data.columns) > 1:
    sns.pairplot(numeric_data, diag_kind='kde', corner=True)
    plt.suptitle('Pair Plot of Numeric Variables', y=1.02)
    plt.show()
else:
    print("Not enough numeric columns for pair plot.")
```
