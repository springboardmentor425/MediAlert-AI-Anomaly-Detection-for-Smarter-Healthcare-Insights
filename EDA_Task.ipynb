```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/Healthcare Providers.csv')
```

```python
data.head()
```

```python
print(data.describe())
```

```python
data.count
print(data.count())
```

```python
print("Missing valuse")
missing_values=data.isnull().sum()
percentages=(missing_values/len(data))*100
dataOfMissingValues=pd.DataFrame({'Missing_Values':missing_values,"Percentages":percentages})
dataOfMissingValues
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Provider Type')  # Adjust 'Provider_Type' based on your dataset
plt.title('Distribution of Provider Types')
plt.xticks(rotation=45)
plt.show()
```

```python
provider_counts = data['Provider Type'].value_counts()
print("Provider Type Counts:\n", provider_counts)

most_common_provider = provider_counts.idxmax()
least_common_provider = provider_counts.idxmin()

print(f"Most common provider type: {most_common_provider}")
print(f"Least common provider type: {least_common_provider}")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Gender of the Provider')  # Adjust 'Gender' based on your dataset
plt.title('Gender Distribution Among Providers')
plt.show()

# Gender balance count
gender_counts = data['Gender of the Provider'].value_counts()
print("Gender Distribution:\n", gender_counts)
```

```python
# Histograms for numerical variables
data.hist(bins=30, figsize=(9,5))
plt.suptitle('Histograms of Numerical Features')
plt.show()

# Boxplots for numerical variables
plt.figure(figsize=(12,6))
sns.boxplot(df=data.select_dtypes(include=['float64', 'int64']))
plt.title('Boxplots of Numerical Features')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select only numerical features for correlation analysis
numerical_data = data.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_data.corr()

# Plotting the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
state_counts = data['State Code of the Provider'].value_counts()
most_providers_state = state_counts.idxmax()
least_providers_state = state_counts.idxmin()

print(f"State with Most Providers: {most_providers_state} ({state_counts.max()} providers)")
print(f"State with Least Providers: {least_providers_state} ({state_counts.min()} providers)")
```

```python
#City-wise Analysis
city_metrics = data.groupby('City of the Provider').agg({
    'National Provider Identifier': 'count',   # Count of providers
    'Number of Services': 'sum',              # Total services
    'Average Medicare Payment Amount': 'mean',  # Average payment cost
    'Average Medicare Allowed Amount': 'mean'   # Average allowed cost
}).reset_index().round(2)
city_metrics.columns = [
    'City',
    'Provider Count',
    'Total Services',
    'Avg Medicare Payment Amount',
    'Avg Medicare Allowed Amount'
]
top_cities_by_providers = city_metrics.sort_values(by='Provider Count', ascending=False).head(5)
print("Top 10 Cities by Most Active Healthcare Providers:")
print(top_cities_by_providers)
```

```python
# Analyzing average submitted charges vs Medicare Payments

data['Charge_Payment_Ratio'] = data['Average Medicare Payment Amount'] / data['Average Medicare Payment Amount']

plt.figure(figsize=(12, 8))
sns.scatterplot(data=data,
                x='Average Medicare Payment Amount',
                y='Average Medicare Payment Amount',
                alpha=0.5)

plt.title('Submitted Charges vs Medicare Payments', fontsize=14)
plt.xlabel('Average Medicare Payment Amount ($)', fontsize=10)
plt.ylabel('Average Medicare Payment Amount ($)', fontsize=10)

max_val = max(data['Average Medicare Payment Amount'].max(),
              data['Average Medicare Payment Amount'].max())
plt.plot([0, max_val], [0, max_val], 'r-', label='Charge = Payment Line')
plt.legend()
plt.tight_layout()
plt.show()
```
